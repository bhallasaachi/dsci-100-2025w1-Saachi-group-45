<h1><center><u>Methods</u></center></h1>

<p>I would use <strong>k-nn classification </strong> to answer my predictive question on the players.csv dataset. This method was chosen because I wish to predict a categorical class (subscribed vs not subscribed) based on 3 observations: age, played_hours, and gender.  Since gender is a categorical variable, I must first create a dummy variable for it to represent it numerically, before I can use it in knn-classification as a predictor variable.  
K-nn classification is non-parametric and does not make assumptions on how the data is distributed. For this method, I must assume that the data reported is accurate such as the self-reported experience level for each player. I must also perform mean imputation to account for N/A entries for my predictor variables. As we see in summary stats, there is a class imbalance of entries with subscribe listed as True vs False (144 vs 52), as such I must oversample the False label in order to avoid True label overepresentation in my prediction.  A potential limitation of the method selected is that there are multiple predictor variable and in a graphical representation, this would look like a 3-D graph which is hard to interpret upon sight. Additionally, knn classification can be slow to proccess data with too many predictive variables. I will use the following steps to process my data: <br>
1)	Convert gender to a factor using the as_factor function as k-nn classification does not work with predictors that are characters.<br>
2)  Create a dummy variable for gender <br>
3)	Use the K-nearest neighbours algorithm included in tidymodels<br>
4)	Select  the columns needed for my classification (I already completed this step in the Data Wrangling section)<br>
6)	Set the seed<br>
7)	Split data into training and testing set with 70% in training set and 30% in testing set (arbitrarily chosen since we are not too worried about being meticulously accurate with data for videogames<br>
8)	Center and scale the data<br>
9)	Split the training set into a training set and a validation set using 5-fold cross-validation (arbitrarily chosen)<br>
10)	Tune the classifier to find the best value of K that gives us the highest accuracy and fit the model using the training data.<br>
11)	Use that value of K in a workflow with predictor values of age, played_hours, and gender to predict the class of subscribe on the test set.</p>
