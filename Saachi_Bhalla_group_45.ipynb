<h1><u>Data Wrangling</u></h1>

<h3>Calculating Summary Stats (mean) on Players Dataset</h3>

```r
players_mean <- players |>
select(Age, played_hours)|>
map(mean, na.rm = TRUE)
players_mean
```

<h3>Calculating distribution of observations within gender and experience variables</h3>

```r
experience_percentage <- nrow(players)
players |>
  group_by(experience) |>
  summarize(
    count = n(),
  )

gender_percentage <- nrow(players)
players |>
  group_by(gender) |>
  summarize(
    count = n(),
  )
```

<h1><u>Visualizing the Data</u></h1>

```r
options(repr.plot.height = 10, repr.plot.width = 17)
age_plot <- players |>
ggplot(aes(x = Age, fill = subscribe)) +
geom_bar(position = "dodge") +
labs(
    x= "Age of Player",
    y= "Number of Players",
    title = "Age vs Played Hours Plot",
    fill = "Subscription Status"
    )+
theme(text = element_text(size = 20))
age_plot
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
gender_plot <- players |>
ggplot(aes(x = gender, fill = subscribe)) +
geom_bar() +
labs(
    x= "Gender of Player",
    y= "Number of Players",
    title = "Gender vs Subscription",
    fill = "Subscription Status"
    )+
theme(axis.text.x = element_text(angle = 45, hjust = 1))+
theme(text = element_text(size = 20))
gender_plot
```

```r
experience_plot <- players |>
ggplot(aes(x = experience, fill = subscribe)) +
geom_bar() +
labs(
    x= "Skill level of Player",
    y = "Number of Players",
    title = "Subscribtion status vs Skill Level",
    fill = "Subscription Status"
    )+
theme(axis.text.x = element_text(angle = 45, hjust = 1))+
theme(text = element_text(size = 20))
experience_plot
```

```r
options(repr.plot.height = 10, repr.plot.width = 17)
Hours_played_plot <- players |>
ggplot(aes(x = played_hours)) +
geom_histogram(binwidth = 5,) +
labs(
       x = "Hours Played",
       y = "Number of Players",
    title = "Distribution of Played Hours")+
theme(text = element_text(size = 20))
Hours_played_plot
```
